## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Wrangling

### Gathering Data

```python
# Membaca setiap data per wilayah dengan menggunakan library pandas
data_Aotizhongxin = pd.read_csv("./data/PRSA_Data_Aotizhongxin_20130301-20170228.csv")
data_Changping = pd.read_csv("./data/PRSA_Data_Changping_20130301-20170228.csv")
data_Dingling = pd.read_csv("./data/PRSA_Data_Dingling_20130301-20170228.csv")
data_Dongsi = pd.read_csv("./data/PRSA_Data_Dongsi_20130301-20170228.csv")
data_Guanyuan = pd.read_csv("./data/PRSA_Data_Guanyuan_20130301-20170228.csv")
data_Gucheng = pd.read_csv("./data/PRSA_Data_Gucheng_20130301-20170228.csv")
data_Huairou = pd.read_csv("./data/PRSA_Data_Huairou_20130301-20170228.csv")
data_Nongzhanguan = pd.read_csv("./data/PRSA_Data_Nongzhanguan_20130301-20170228.csv")
data_Shunyi = pd.read_csv("./data/PRSA_Data_Shunyi_20130301-20170228.csv")
data_Tiantan = pd.read_csv("./data/PRSA_Data_Tiantan_20130301-20170228.csv")
data_Wanliu = pd.read_csv("./data/PRSA_Data_Wanliu_20130301-20170228.csv")
data_Wanshouxigong = pd.read_csv("./data/PRSA_Data_Wanshouxigong_20130301-20170228.csv")

data_Aotizhongxin.head(5)
```

```python
data_Aotizhongxin.info()
data_Aotizhongxin.describe()
```

**Insight:**
- Dilakukan impor data untuk setiap wilayah / kota
- Didapatkan kira-kira 35.000 data untuk setiap wilayah

### Assessing Data

```python
# Menggabungkan data dari setiap wilayah
combined_df = pd.concat([data_Aotizhongxin, data_Changping, data_Dingling, data_Dongsi,
                         data_Guanyuan, data_Gucheng, data_Huairou, data_Nongzhanguan,
                         data_Shunyi, data_Tiantan, data_Wanliu, data_Wanshouxigong],
                        ignore_index=True)
```

```python
# Info dari penggabungan data beserta tipe datanya
combined_df.info()
```

```python
# Memeriksa apakah ada missing values
combined_df.isnull().sum()
```

```python
# Hitung Interquartile Range / IQR untuk data PM10
q25, q75 = combined_df['PM10'].quantile(0.25), combined_df['PM10'].quantile(0.75)
IQR = q75 - q25

# Tentukan batas untuk outlier
cut_off = IQR * 1.5
minimum, maximum = q25 - cut_off, q75 + cut_off

# Temukan outlier
outliers = combined_df[(combined_df['PM10'] < minimum) | (combined_df['PM10'] > maximum)]

# Tampilkan jumlah data outlier dan contoh 5 data dari outlier
print(f"Jumlah outlier: {outliers.shape[0]}")
outliers.head()
```

**Insight:**
- Ditemukan adanya missing values pada field PM10 sebanyak 6 ribu data
- Teridentifikasi adanya outlier / data yang diluar jangkauan sebanyak 14 ribu data

### Cleaning Data

```python
# Menghapus data yang value PM10-nya tidak ada / null
print(f'Jumlah data sebelum = {combined_df.shape[0]}')
cleaned_df = combined_df.dropna(subset=['PM10'])
print(f'Jumlah data sesudah = {cleaned_df.shape[0]}')
```

```python
# Menghapus data outlier
lower_t = cleaned_df['PM10'] < minimum
more_t = cleaned_df['PM10'] > maximum

cleaned_df.drop(cleaned_df[lower_t].index)
cleaned_df.drop(cleaned_df[more_t].index)

print(f'Jumlah data setelah outlier dihapus = {cleaned_df.shape[0]}')
```

```python
# Menyimpan hasil data yang telah dibersihkan
cleaned_df.to_csv('./dashboard/main_data.csv', index=False)
```

**Insight:**
- Dilakukan penghapusan data yang null dan data outlier
- Menyimpan hasil data wrangling ke dalam csv baru

## Exploratory Data Analysis (EDA)

### Explore tren polusi udara (PM10) di setiap wilayah dalam dataset dari tahun 2013-2017

```python
# Membuat kolom baru, yaitu 'datetime' untuk analisis berdasarkan waktu
cleaned_df = cleaned_df.copy()
cleaned_df['datetime'] = pd.to_datetime(cleaned_df[['year', 'month', 'day', 'hour']])

# Grouping berdasarkan bulan dan tahun untuk menghitung rata-rata PM10 di setiap wilayah
monthly_pm10 = cleaned_df.groupby([cleaned_df['datetime'].dt.to_period('M'), 'station'])['PM10'].mean().unstack()

print(monthly_pm10.head())
```

### Explore waktu konsentrasi PM10 yang paling tinggi dari setiap wilayah


```python
# Menghitung rata-rata PM10 berdasarkan jam
hourly_pm10 = cleaned_df.groupby(['hour', 'station'])['PM10'].mean().unstack()

print(hourly_pm10.head())
```

**Insight:**
- Dilakukan grouping untuk menjawab dari pertanyaan analisis

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Visualisasi tren PM10 dari tahun 2013 hingga 2017 di setiap wilayah
plt.figure(figsize=(14, 7))
monthly_pm10.plot(marker='o', linestyle='-', colormap='viridis')
plt.title('Tren Rata-rata Konsentrasi PM10 di Setiap Wilayah (2013-2017)')
plt.xlabel('Tanggal')
plt.ylabel('Rata-rata Konsentrasi PM10 (µg/m³)')
plt.legend(title='Wilayah', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
# Visualisasi konsentrasi PM10 berdasarkan jam
plt.figure(figsize=(28, 14))
hourly_pm10.plot(marker='o', linestyle='-', colormap='plasma')
plt.title('Rata-rata Konsentrasi PM10 Berdasarkan Jam di Setiap Wilayah')
plt.xlabel('Jam dalam Sehari')
plt.ylabel('Rata-rata Konsentrasi PM10 (µg/m³)')
plt.legend(title='Wilayah', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Insight:**
- Dilakukan visualisasi dari data yang sudah di proses untuk masing-masing pertanyaan analisis

## Conclusion

- Untuk tren dari polusi udara, khususnya faktor PM10, terlihat adanya fluktuasi musiman. Adanya peningkatan konsentrasi PM10 ketika musim dingin, bisa disebabkan oleh faktor diperlukannya penggunaan bahan bakar untuk menyalakan pemanas. Untuk beberapa wilayah terlihat cukup tinggi dapat disebabkan faktor geologis dan adanya kegiatan industri yang cukup padat.
- Untuk pola konsentrasi PM10 yang dilihat dari jam di setiap wilayahnya terlihat bahwa pagi sekitar pukul 06:00 hingga 10:00 serta sore hingga malam hari sekitar pukul 17:00 hingga 22:00 ada peningkatan konsentrasi PM10. Dapat disebabkan oleh aktivitas manusia atau area industrial, dimana lalu lintas padat karena jam pulang pergi kerja.
